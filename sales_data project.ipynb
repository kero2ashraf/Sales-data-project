# Let`s make some visualizations

```python
plt.boxplot(df["Salary"]);
plt.title("Salary")
```

```python
duplicates = df.duplicated()
print("Number of duplicates : ",duplicates.sum())

df.drop_duplicates(inplace=True)
```

```python
# Handling inconsistent or incorrect data
df.boxplot(column="Salary")

# Identify the threshold value for outliers
threshold = df["Salary"].mean() + 3 * df["Salary"].std()

# Remove rows with salaries above the threshold
df = df[df["Salary"] <= threshold]
```

```python
# Visualization of numerical variables
numerical_vars = ["Age", "Years of Experience", "Salary"]
df[numerical_vars].hist(bins=10, figsize=(10, 6))
plt.suptitle("Histograms of Numerical Variables", fontsize=14)
plt.show()
```

```python
df.boxplot(column=numerical_vars, figsize=(10, 6))
plt.title("Boxplots of Numerical Variables", fontsize=14)
plt.show()
```

```python
df["Gender"].value_counts().plot(kind="bar", figsize=(10,6))
plt.title("Frequences for Gender" , fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar plot of Education Level
df["Education Level"].value_counts().plot(kind="bar", figsize=(10,15))
plt.title("Frequencies for Education Level",fontsize=14)
plt.xlabel("Education Level")
plt.ylabel("Frequency")
plt.show()
```

```python
#Grouping similar jobs titles into broader categories

job_title_mapping = {
    "Data Scientist" : "Data Science",
    "Data Analyst" : "Data Science",
    "Machine Learning Engineer" : "Data Science",
    "Software Engineer" : "Software Development",
    "Web Developer" : "Software Development",
    "Product Manager": "Product Management",
}

df["Job Title Group"] = df["Job Title"].map(job_title_mapping)

#frequencies for grouped job titles
job_title_group_frequencies = df["Job Title Group"].value_counts(normalize=True) * 100
print("Frequences for grouped job titles : ")
print(job_title_group_frequencies)
```

```python
#bar plot of  grouped job titles
df["Job Title Group"].value_counts().plot(kind="bar", figsize=(10,19))
plt.title("Bar Plot of Job Title Groups", fontsize=14)
plt.xlabel("Job Title Group")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatter plot : Age vs Salary
plt.figure(figsize=(10,6))
plt.scatter(df["Age"],df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# Scatter plot: Experience vs. Salary
plt.figure(figsize=(10, 6))
plt.scatter(df["Years of Experience"], df["Salary"])
plt.title("Experience vs. Salary", fontsize=14)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
# Pie chart: Distribution of Education Level
plt.figure(figsize=(10 ,6))
df["Education Level"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of Education Level", fontsize=14)
plt.ylabel("")
plt.show()
```

```python
# Heatmap: Correlation matrix
corr_matrix = df[["Age", "Years of Experience", "Salary"]].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix", fontsize=14)
plt.show()
```
